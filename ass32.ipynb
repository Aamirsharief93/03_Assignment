```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from CSV
#df = pd.read_csv('your_dataset.csv')  # Replace 'your_dataset.csv' with your actual file name

# Extracting data
years = df['Year']
highest_temps = df['Highest Temp.']
lowest_temps = df['Lowest Temp']
bar_width=0.4

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(years-bar_width, highest_temps,width=bar_width, color='orange', label='Highest Temperature')
plt.bar(years, lowest_temps,width=bar_width, color='blue', label='Lowest Temperature', alpha=0.7)

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.title('Highest and Lowest Temperatures Over the Years')
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from CSV
#df = pd.read_csv('your_dataset.csv')  # Replace 'your_dataset.csv' with your actual file name

# Extracting data
highest_temps = df['Highest Temp.']
monthly_precipitation = df['Monthly Precipatation(cm)']
water_scarcity_percent = df['Percent of People reporting Water Scarcity']
heat_stroke_deaths = df['People dying from Heat Strokes']

# Scatter plot: Highest Temp vs Monthly Precipitation
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(highest_temps, monthly_precipitation, color='blue', alpha=0.7)
plt.title('Highest Temp vs Monthly Precipitation')
plt.xlabel('Highest Temperature (°C)')
plt.ylabel('Monthly Precipitation(cm)')

# Scatter plot: Highest Temp vs Percent of People reporting Water Scarcity
plt.subplot(1, 2, 2)
plt.bar(highest_temps, water_scarcity_percent, color='green', alpha=0.7)
plt.title('Highest Temp vs Water Scarcity Percentage')
plt.xlabel('Highest Temperature(°C)')
plt.ylabel('Percent of People reporting Water Scarcity')

plt.tight_layout()

# Bar plot: People dying from Heat Strokes
plt.figure(figsize=(10, 6))
plt.bar(df['Year'], heat_stroke_deaths, color='red', label='Heat Stroke Deaths')
plt.title('Heat Stroke Deaths Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.legend()

# Show the plots
plt.show()
```

```python
c1 = 'Highest Temp.'    
c2 = 'Lowest Temp'
c3 = 'People dying from Heat Strokes'
c4 = 'Monthly Precipatation(cm)'
c5 = 'Percent of People reporting Water Scarcity'
c6 = 'Aridity Index'
c7 = 'Paddy Production(Tonnes)'
c8 = 'Vegerable Production(Tonns)'

temp_variables = [c1, c2]
other_variables = [c3, c4, c5, c6, c7, c8]

fig, axis = plt.subplots(len(other_variables), len(temp_variables), figsize=(20, 40))

for i, other_var in enumerate(other_variables):
    for j, temp_var in enumerate(temp_variables):
        axis[i, j].bar(df[temp_var], df[other_var], color='blue')  # Corrected color parameter
        axis[i, j].set_title(f'{temp_var} vs {other_var}')
        axis[i, j].set_xlabel(temp_var)
        axis[i, j].set_ylabel(other_var)

# Adjust layout to prevent overlapping titles
plt.tight_layout()

# Show the plots
plt.show()
```
